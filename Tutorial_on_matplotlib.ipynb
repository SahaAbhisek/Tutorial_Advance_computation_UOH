## Stream Plots

-3<x,y<3,  n=100
$$ u = -1 - x^{2} +y $$
$$ v = 1 + x - y^{2} $$


```python
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

xv, yv = np.meshgrid(x,y)
u = -1 - xv**2 + yv
v = 1 + xv - yv**2

vel = np.sqrt(u**2+ v**2)
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.streamplot(x, y, u, v)

plt.subplot(2,2,2)
plt.streamplot(x, y, u,v, color=vel)
plt.colorbar()

plt.subplot(2,2,3)
lw = 5*vel/vel.max()
plt.streamplot(x, y, u,v, linewidth=lw)

plt.subplot(2,2,4)
plt.streamplot(x, y, u,v, start_points=[(0,1), (2,1)])

plt.show()
```

## Question 1

Let $f(x,y) = e^{-(x^2+y^2)} \cdot \sin(x)$ for $-2 \leq x \leq 2$ and $-2 \leq y \leq 2$

1. Make a contour plot of $f$
2. Find the volume $|f(x,y)|$ in the specified $x$ and $y$ range
3. Find the volume $|f(x,y)|$ only in the region where $\sqrt{x^2+y^2}>0.5$ 

```python
def f(x,y):
    return np.exp(-x**2-y**2)*np.sin(x)
```

```python
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)

X, Y = np.meshgrid(x,y)

Z = f(X,Y)
```

```python
plt.contourf(x,y,Z, levels=15)
plt.colorbar()
```

```python
np.sum(np.abs(Z.ravel()) * np.diff(x)[0] * np.diff(y)[0])
```

```python
g = Z[X**2 + Y**2>0.5**2]
np.sum(np.abs(g.ravel()) * np.diff(x)[0] * np.diff(y)[0])
```
